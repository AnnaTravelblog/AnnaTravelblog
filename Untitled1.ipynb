```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Загрузка данных
data = pd.read_csv('london_weekdays.csv')

# Построение боксплота для визуализации распределения стоимости аренды в зависимости от типа комнаты
plt.figure(figsize=(10, 6))
sns.boxplot(x='room_type', y='realSum', data=data)
plt.title('Distribution of Rental Prices by Room Type')
plt.xlabel('Room Type')
plt.ylabel('Rental Price (realSum)')
plt.show()

# Проверка значимости различий с использованием теста ANOVA
anova_result = stats.f_oneway(
    data[data['room_type'] == 'Entire home/apt']['realSum'],
    data[data['room_type'] == 'Private room']['realSum'],
    data[data['room_type'] == 'Shared room']['realSum']
)

print(f'ANOVA result: F={anova_result.statistic}, p={anova_result.pvalue}')
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

# Load the specific file for London weekdays
file_path = 'london_weekdays.csv'
london_weekdays_df = pd.read_csv(file_path)

# Display the first few rows of the dataframe
print(london_weekdays_df.head())

# Check for missing values
missing_values = london_weekdays_df.isnull().sum()
print(missing_values)

# Describe the distribution of variables
distribution = london_weekdays_df.describe()
print(distribution)
```

```python
# Visualizing the distribution of rental prices in London on weekdays
plt.figure(figsize=(10, 6))
sns.histplot(london_weekdays_df['realSum'], bins=50, kde=True)
plt.title('Distribution of Rental Prices in London on Weekdays')
plt.xlabel('Rental Price (realSum)')
plt.ylabel('Frequency')
plt.show()

# Visualizing the distribution of guest satisfaction in London on weekdays
plt.figure(figsize=(10, 6))
sns.histplot(london_weekdays_df['guest_satisfaction_overall'], bins=50, kde=True)
plt.title('Distribution of Guest Satisfaction in London on Weekdays')
plt.xlabel('Guest Satisfaction Overall')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualizing the distribution of rental prices by room type in London on weekdays
plt.figure(figsize=(10, 6))
sns.boxplot(x='room_type', y='realSum', data=london_weekdays_df)
plt.title('Distribution of Rental Prices by Room Type in London on Weekdays')
plt.xlabel('Room Type')
plt.ylabel('Rental Price (realSum)')
plt.show()

# Performing ANOVA test to check for significant differences
anova_result = stats.f_oneway(
    london_weekdays_df[london_weekdays_df['room_type'] == 'Entire home/apt']['realSum'],
    london_weekdays_df[london_weekdays_df['room_type'] == 'Private room']['realSum'],
    london_weekdays_df[london_weekdays_df['room_type'] == 'Shared room']['realSum']
)

print(f'ANOVA result: F={anova_result.statistic}, p={anova_result.pvalue}')

```
