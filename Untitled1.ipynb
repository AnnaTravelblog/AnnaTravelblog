```python
TOKEN = '6730662155:AAGsVZkU4AyA7DV6oJKto0nzAZZmnfcSFKk'
OPENWEATHERMAP_API_KEY = 'f98028a3cd35a770ee7e30d447f45301'

BASE_URL = f'https://api.telegram.org/bot{TOKEN}/'
WEATHER_API_URL = 'https://api.openweathermap.org/data/2.5/weather?'
# Приветственные сообщения

GREETINGS = [
    "Привет! Я твой путешественник-бот. Готов помочь с планированием твоих следующих приключений!",
    "Здравствуй! Я готов отправиться с тобой в путешествие!",
    "Приветствую тебя, путешественник! Куда собираешься отправиться?",
    "Привет! Я здесь, чтобы помочь тебе с выбором следующего направления для путешествия."
]



nlp = spacy.load("xx_ent_wiki_sm")



DESTINATIONS = {
    "париж": {
        "description": "Париж - прекрасный город! Обязательно посети Эйфелеву башню и Лувр."
    },
    "токио": {
        "description": "Токио - удивительный город! Не пропусти Цукадзи и Сенсодзи."
    },
    "рим": {
        "description": "Рим - великолепный город! Посети Колизей и Ватикан."
    },

     "лондон": {
        "description": "Лондон - столица Великобритании! Посети Биг Бен и Британский музей."
    },
    "барселона": {
        "description": "Барселона - город архитектурных шедевров! Посети собор Святого Семейства и Парк Гуэль."
    },
    "нью-йорк": {
        "description": "Нью-Йорк - город, который никогда не спит! Открой для себя Манхэттен и Центральный парк."
    },
    "москва": {
        "description": "Москва - столица России! Посети Красную площадь и Кремль."
    },
    "киев": {
        "description": "Киев - древний город с богатой историей! Посети Софиевский собор и Подол."
    },
    "бангкок": {
        "description": "Бангкок - экзотический город в сердце Таиланда! Наслаждайся вкусной тайской кухней и посети храм Ват Арун."
    },
    "прага": {
        "description": "Прага - город старинных улочек и замков! Посети Карлов мост и Пражский град."
    },
    "дубай": {
        "description": "Дубай - город впечатляющих небоскребов и роскоши! Посети Бурдж Халифа и Дубай Молл."
    },
    "сидней": {
        "description": "Сидней - живописный город на побережье! Посети оперный театр и Сиднейский харбор."
    },
    "амстердам": {
        "description": "Амстердам - город каналов и великолепной архитектуры! Посети музей Ван Гога и Дом Анны Франк."
    },
    "флоренция": {
        "description": "Флоренция - исторический город с богатым культурным наследием! Посети Галерею Уффици и собор Санта-Мария-дель-Фьоре."
    },
    "киото": {
        "description": "Киото - древняя столица Японии с множеством храмов и садов! Посети Золотой павильон и храм Фушими Инари."
    },
    "сан-франциско": {
        "description": "Сан-Франциско - культурный и технологический центр США! Посети знаменитый мост Голден Гейт и остров Алькатрас."
    },
    "шанхай": {
        "description": "Шанхай - современный мегаполис с высокими небоскребами! Посети набережную Бунд и Ципао."
    },
    "кипр": {
        "description": "Кипр - остров любви и истории! Посети археологический парк Пафоса и монастырь Киккос."
    },
    "рио-де-жанейро": {
        "description": "Рио-де-Жанейро - город пляжей и карнавалов! Посети Христа-Искупителя и пляж Ипанема."
    },
    "сеул": {
        "description": "Сеул - космополитичный город с высокими технологиями! Посети дворец Гёнбоккун и Торговый центр Дунгдаемун."
    },
    "будапешт": {
        "description": "Будапешт - город с тысячелетней историей и термальными источниками! Посети Рыбацкий бастион и парламент."
    },
    "каир": {
        "description": "Каир - древний город с пирамидами и сфинксами! Посети пирамиду Хеопса и музей египетской античности."
    },
    "санкт-петербург": {
        "description": "Санкт-Петербург - северная столица России с богатым культурным наследием! Посети Эрмитаж и Петергоф."
    },
    "мельбурн": {
        "description": "Мельбурн - культурный и спортивный центр Австралии! Посети здание парламента и Royal Botanic Gardens."
    },
     "дублин": {
        "description": "Дублин - живописный город с историческими замками и пабами! Посети Замок Дублин и бар Темпл Бар."
    },
    "вена": {
        "description": "Вена - столица Австрии с красивой архитектурой и классической музыкой! Посети дворец Шенбрунн и кафе Централь."
    },
    "кишинев": {
        "description": "Кишинёв - столица Молдовы! Посети Национальный музей и Старую крепость."
    },
    "тирасполь": {
        "description": "Тирасполь - столица Приднестровской Молдавской Республики! Посети Приднестровский государственный университет и Парк Славы."
    },
    "кипр": {
        "description": "Кипр - остров любви и истории! Посети археологический парк Пафоса и монастырь Киккос."
    },
    "рио-де-жанейро": {
        "description": "Рио-де-Жанейро - город пляжей и карнавалов! Посети Христа-Искупителя и пляж Ипанема."
    },
    "сеул": {
        "description": "Сеул - космополитичный город с высокими технологиями! Посети дворец Гёнбоккун и Торговый центр Дунгдаемун."
    },
}

def handle_message(update: Update, context: CallbackContext) -> None:
    text = update.message.text.lower()
    if "спасибо" in text:
        send_message(update.message.chat_id, "Пожалуйста!")
    elif "всегда пожалуйста" in text:
        send_message(update.message.chat_id, "Рад помочь!")
    else:
        for destination in DESTINATIONS:
            if destination in text:
                description = DESTINATIONS[destination]["description"]
                weather_info = get_weather_info(destination)
                send_message(update.message.chat_id, f"{description}\n\n{weather_info}")
                return
        send_message(update.message.chat_id, "Я не знаю такого места. Может быть, выбери что-то другое?")

def get_weather_info(city: str) -> str:
    params = {'q': city, 'appid': OPENWEATHERMAP_API_KEY, 'units': 'metric', 'lang': 'ru'}
    response = requests.get(WEATHER_API_URL, params=params)
    if response.status_code == 200:
        data = response.json()
        weather_description = data['weather'][0]['description']
        temperature = data['main']['temp']
        return f"Погода в {city.capitalize()}: {weather_description}, Температура: {temperature}°C"
    else:
        return "Информация о погоде недоступна."

def send_message(chat_id, text):
    data = {
        'chat_id': chat_id,
        'text': text
    }
    response = requests.post(BASE_URL + 'sendMessage', data=data)
    print(response.json())

def start(update, context):
    update.message.reply_text(random.choice(GREETINGS))


def main():
    updater = Updater(TOKEN)
    dispatcher = updater.dispatcher

    # Обработчик команды /start
    dispatcher.add_handler(CommandHandler("start", start))

    # Добавление обработчика для обычных текстовых сообщений
    dispatcher.add_handler(MessageHandler(Filters.text & ~Filters.command, handle_message))

    updater.start_polling()
    updater.idle()

if __name__ == '__main__':
    main()
```
